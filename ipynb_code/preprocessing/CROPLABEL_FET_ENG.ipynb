```python
import numpy as np
import pandas as pd

# for plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# feature scaling
from sklearn.preprocessing import MinMaxScaler

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
FinDaDr = pd.read_csv('C:/Users/hp/FIN_DATA/Final_Data_Draft.csv')
FinDaDr.head()
```

```python
vars = [var for var in FinDaDr.columns]
```

```python

scaler = MinMaxScaler(feature_range=(0,4))
dc1 = FinDaDr[(FinDaDr['Crop'] == 1)].dropna(how='any',axis=0,subset=['Yield'])
sdc1 = scaler.fit_transform(dc1[vars])
sdc1 = np.around(sdc1)
##########################################################################################

scaler = MinMaxScaler(feature_range=(0,4))
dc2 = FinDaDr[(FinDaDr['Crop'] == 2)].dropna(how='any',axis=0,subset=['Yield'])
sdc2 = scaler.fit_transform(dc2[vars])
sdc2 = np.around(sdc2)
##########################################################################################

scaler = MinMaxScaler(feature_range=(0,4))
dc3 = FinDaDr[(FinDaDr['Crop'] == 3)].dropna(how='any',axis=0,subset=['Yield'])
sdc3 = scaler.fit_transform(dc3[vars])
sdc3 = np.around(sdc3)
##########################################################################################

scaler = MinMaxScaler(feature_range=(0,4))
dc4 = FinDaDr[(FinDaDr['Crop'] == 4)].dropna(how='any',axis=0,subset=['Yield'])
sdc4 = scaler.fit_transform(dc4[vars])
sdc4 = np.around(sdc4)
##########################################################################################

scaler = MinMaxScaler(feature_range=(0,4))
dc5 = FinDaDr[(FinDaDr['Crop'] == 5)].dropna(how='any',axis=0,subset=['Yield'])
sdc5 = scaler.fit_transform(dc5[vars])
sdc5 = np.around(sdc5)
##########################################################################################

scaler = MinMaxScaler(feature_range=(0,4))
dc6 = FinDaDr[(FinDaDr['Crop'] == 6)].dropna(how='any',axis=0,subset=['Yield'])
sdc6 = scaler.fit_transform(dc6[vars])
sdc6 = np.around(sdc6)
##########################################################################################

scaler = MinMaxScaler(feature_range=(0,4))
dc7 = FinDaDr[(FinDaDr['Crop'] == 7)].dropna(how='any',axis=0,subset=['Yield'])
sdc7 = scaler.fit_transform(dc7[vars])
sdc7 = np.around(sdc7)
##########################################################################################

scaler = MinMaxScaler(feature_range=(0,4))
dc8 = FinDaDr[(FinDaDr['Crop'] == 8)].dropna(how='any',axis=0,subset=['Yield'])
sdc8 = scaler.fit_transform(dc8[vars])
sdc8 = np.around(sdc8)
##########################################################################################
```

```python
sdcFra1 = pd.DataFrame(sdc1, columns = FinDaDr.columns)
sdcFra2 = pd.DataFrame(sdc2, columns = FinDaDr.columns)
sdcFra3 = pd.DataFrame(sdc3, columns = FinDaDr.columns)
sdcFra4 = pd.DataFrame(sdc4, columns = FinDaDr.columns)
sdcFra5 = pd.DataFrame(sdc5, columns = FinDaDr.columns)
sdcFra6 = pd.DataFrame(sdc6, columns = FinDaDr.columns)
sdcFra7 = pd.DataFrame(sdc7, columns = FinDaDr.columns)
sdcFra8 = pd.DataFrame(sdc8, columns = FinDaDr.columns)
```

```python
sdcFra1 = sdcFra1.append(sdcFra2)
sdcFra1 = sdcFra1.append(sdcFra3)
sdcFra1 = sdcFra1.append(sdcFra4)
sdcFra1 = sdcFra1.append(sdcFra5)
sdcFra1 = sdcFra1.append(sdcFra6)
sdcFra1 = sdcFra1.append(sdcFra7)
sdcFra1 = sdcFra1.append(sdcFra8)
```

```python
sdcFra1 = sdcFra1.reset_index()
```

```python
FinDaDr['CropLabel'] = sdcFra1['Yield']
```

```python
FinDaDr = FinDaDr.drop(columns = ['Yield'])
```

```python
# let me show you what I mean by monotonic relationship between labels and target
def analyse_vars(df, var):
    df = df.copy()
    df.groupby(var)['Production'].median().plot.bar()
    plt.title(var)
    plt.ylabel('Production')
    plt.show()
    

analyse_vars(FinDaDr, 'CropLabel')
```
